```python
#Q:Count the number of countries for which the total_tests data is missing.
```

```python
covid_data=pd.read_csv('covid-countries-data.csv')
```

```python
covid_data['total_tests'].isna().sum()
```

```python
merged_df= df.merge(covid_data, on='location')
```

```python
merged_df['tests_per_million'] = merged_df['total_tests'] * 1e6 / merged_df['population']
merged_df['cases_per_million'] = merged_df['total_cases'] * 1e6 / merged_df['population']
merged_df['deaths_per_million'] = merged_df['total_deaths'] * 1e6 / merged_df['population']
```

```python
#Q: Create a dataframe with 10 countires that have highest number of tests per million people.
```

```python
top10 = merged_df[['tests_per_million','location']].sort_values('tests_per_million', ascending=False)
```

```python
top10=top10.head(10)
```

```python
#Q: Create a dataframe with 10 countires that have highest number of positive cases per million people.
```

```python
top10poscases = merged_df[['cases_per_million','location']].sort_values('cases_per_million', ascending=False)
```

```python
top10poscases=top10poscases.head(10)
```

```python
 #Q: Count number of countries that feature in both the lists "20 countries with lowest GDP per capita" and "20 countries with the lowest number of hospital beds per thousand population". Only consider countries with a population higher than 10 million while creating the list.
```

```python
lowest_GDP= df.sort_values('gdp_per_capita', ascending= True)
```

```python
lowest_GDP.head(20)
```

```python
lowest_hospital_beds= df.sort_values('hospital_beds_per_thousand', ascending= True).head(20)
```

```python
lowest_hospital_beds
```

```python
final = set(lowest_GDP['location']).intersection(set(lowest_hospital_beds['location']))
```

```python
filtered_GDP = lowest_GDP[lowest_GDP['population'] > 10e6]

filtered_hospital_beds = lowest_hospital_beds[lowest_hospital_beds['population'] > 10e6]


common_countries = set(filtered_GDP['location']).intersection(set(filtered_hospital_beds['location']))


num_common_countries = len(common_countries)

print("Number of countries appearing in both lists with populations higher than 10 million:", num_common_countries)
```
